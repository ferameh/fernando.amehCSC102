```python
#Welcome to my mathematics class.
# I will give you an ILLUSTRATION on how cubic equation works.
# In a cubic equation. It is written like this: ax^3 + bx^2 + cx + d = 0
# where a, b, c and d are the coeffiecients. Oops for non- lovers of mathematics i.e. coefficients are the clothes worn on the body.
# I do hope that it's clear enough.
a = float(input("Enter any figure:"))
b = float(input("Enter a figure:"))
c = float(input("Enter a figure:"))
d = float(input("Enter a figure: "))

#We are going to divide through by the co-efficient of x^3
# In other word. I will let you know what is a1, a2 and a3 e.t.c.
# Below is exactly what I jotted in the last two comment above.
a1 = b / a
print("a1=",a1)
a2 = c / a
print("a2",a2)
a3 = d / a
print("a3",a3)

# What I'm about to do next is quite complicated. I will recommend First layout Mathematics textbook for beginners  by Dr. Ameh Fernando.
# I'm going  to write out how to get Q, R, S, T.

# Q = (3a2 - a1^2)/9.... So from here, we will declare a variable for the numerator. A numerator is the number on top of the fraction.
am = 3*a2 - a1**2
# Recall  Q = (3a2 - a1^2)/9....
Q = am /9
print("Q:",Q)

# We are making progress. Don't giveup our goal is to find the roots of the cubic equation.
# We are going to find R

# R = (9*a1*a2 - 27a3 - 2a1^3)
# Let eh REPRESENT 9*a2*a1 - 27*a3 - 2*a1**3
eh = 9*a2*a1 - 27*a3 - 2*a1**3
R = eh/54
print("R:",R)

# S = (R + (Q^3 + R^3)**1/2)**1/3
# Work with me with the above formular.

pe = (Q**3 + R**2)**1/2
S = (R + pe)**1/3
print("S:",S)

# T = (R - (Q^3 + R^2)^1/2)^1/3
T = (R - pe)**1/3
print("T:", T)

# Good job. Let's round up our final session. 
# In a cubic equation; we are expected to find three roots. They are x1, x2 and x3.
#......X1 = S + T - (1/3)*a1

X1 = S + T - (1/3)*a1

print("X1:",X1)
x2 = -1/2*(S + T) - 1/3*a1 + 1/2*(-1**1/2)*(3)**1/3*(S-T) # Where i is sqrt of -1
print("x2:",x2)

X3 = -1/2*(S + T) - 1/3*a1 - 1/2*(-1**1/2)*(3)**1/3*(S-T) 
print("X3:", X3)

```
